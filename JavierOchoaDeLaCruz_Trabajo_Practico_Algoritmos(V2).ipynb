## 3 ¿Cual es la funcion objetiva ?

funcion objetivo:
## MAXIMIZAR ##
    
La función objetivo en este problema es maximizar la audiencia total. Dicha audiencia se calcula sumando la audiencia de cada partido multiplicada por el coeficiente correspondiente al horario en el que se juega dicho partido y considerando las posibles reducciones de audiencia en caso de que haya coincidencia de partidos.
$$funcion maxima (\sum_{i=1}^{N}audiencia(i)* PondHorario(i) * reduccion_por_coincidencias(i) ) $$

Donde:
* `N` Es el número de partidos
* `audiencia` Es la audiencia del partido ,que depende de las categorías de los equipos enfrentados.
* `PondHorario` Es coeficiente de ponderación del horario en el que se juega el partido
* `reduccion_por_coincidencias` es el porcentaje de reducción de la audiencia que ajusta debido a la coincidencia de horarios con otros partidos ,Es el coeficiente de reducción de audiencia.


La función objetivo busca encontrar la asignación de partidos a horarios que maximice esta suma total de audiencias, sujeta a las restricciones de horarios disponibles y la obligatoriedad de asignar un partido el viernes y otro el lunes.

## ALGORITMO VORAZ ** RESULTADO LLEGAMOS A UN APROX DE 6,5 MM QUE PEUDE SER UN OPTIMO LOCAL

```python

import pandas as pd


# Definición de audiencias base en el nuevo formato
audiencias_base = pd.Series({'AA': 2, 'AB':1.3, 'AC':1, 'BB':0.9, 'BC': 0.75, 'CC':0.47, 'BA':1.3, 'CA':1, 'CB':0.75})

ponderaciones_horarios = pd.Series({'V20h':0.4, 'S12h':0.55, 'S16h':0.7, 'S18h':0.8, 'S20h': 1, 'D12h':0.45, 'D16h':0.75, 'D18h':0.85, 'D20h':1, 'L20h':0.4})
#display(ponderaciones_horarios)

reduccion_por_coincidencias2 =  pd.DataFrame([0, 0.25, 0.45, 0.6, 0.7, 0.75, 0.78, 0.8, 0.8], columns=['%'])
#display(reduccion_por_coincidencias)

nombres = ['Real Madrid','R.Sociedad','Barcelona','Celta','Valencia','Athletic','Villarreal','Alavés','Levante',
           'Espanyol','Sevilla','Betis','Atlético','Getafe', 'Mallorca','Eibar','Leganés','Osasuna','Granada','Valladolid' ]
equipos_disponibles = ['A', 'A', 'A','B','B','B','B','B','B','B','B','B','B','B','C','C','C','C','C','C']
equipos = pd.DataFrame(equipos_disponibles, index = nombres, columns=['Categoría'])
#display(equipos)

#Partidos de la jornada.
partidos_jornada = [['Celta', 'Real Madrid'], ['Valencia', 'R.Sociedad'], ['Mallorca', 'Eibar'],
                    ['Athletic', 'Barcelona'], ['Leganés', 'Osasuna'], ['Villarreal', 'Granada'],
                    ['Alavés', 'Levante'], ['Espanyol', 'Sevilla'], ['Betis', 'Valladolid'], ['Atlético', 'Getafe']]

##************************************************************************************************************************
#Ordenar partidos por audiencias
import numpy as np
def ordenar_partidos_audiencia(equipos_partidos, equipos, pond_cat_partidos):
  audiencia = {}
  for i in range(len(equipos_partidos)):
    aux = equipos_partidos[i]
    catego = str(equipos.loc[aux[0]][0])+str(equipos.loc[aux[1]][0])
    name_partido = str(aux[0])+'-'+str(aux[1])
    audiencia[name_partido] = pond_cat_partidos.loc[catego]
  audiencia = pd.Series(audiencia).sort_values(ascending=True)
  return audiencia


partidos_ordenados = ordenar_partidos_audiencia(partidos_jornada, equipos, audiencias_base)
print(partidos_ordenados)


##*************************************************************************************************************************
horarios_asignados = pd.Series({'V20h':0, 'S12h':0, 'S16h':0, 'S18h':0, 'S20h': 0, 'D12h':0, 'D16h':0, 'D18h':0, 'D20h':0, 'L20h':0})
audiencias_finales = pd.DataFrame(columns=['Horario', 'Audiencia'])
audiencia_tot=0
cont = 0
mejor_audiencia = 0
mejor_lunes = 0
mejor_viernes = 0

for partido, audiencia in zip(partidos_ordenados.index, partidos_ordenados.values):
  mejor_audiencia = 0
  num_coinci = 0
  #Para cada partido se selecciona el mejor horario (el que más audiencia tiene)
  #Aseguramos que al horario del lunes y del viernes se les asigna un partido
  if horarios_asignados.loc['V20h'] ==0:
    mejor_horario = 'V20h'
    num_coinci = horarios_asignados['V20h']
    mejor_audiencia = ponderaciones_horarios['V20h']*audiencia*(1- reduccion_por_coincidencias2.iloc[num_coinci, 0])
  elif horarios_asignados.loc['L20h'] ==0:
    mejor_horario = 'L20h'
    num_coinci = horarios_asignados['L20h']
    mejor_audiencia = ponderaciones_horarios['L20h']*audiencia*(1- reduccion_por_coincidencias2.iloc[num_coinci, 0])
  else: #Para el resto de partidos seleccionamos el horario que maximice la audiencia
    for horario in ponderaciones_horarios.index:
      audiencia_actual = ponderaciones_horarios[horario]*audiencia*(1- reduccion_por_coincidencias2.iloc[horarios_asignados[horario], 0])

      if audiencia_actual.astype(float) > mejor_audiencia:
        mejor_audiencia = audiencia_actual
        mejor_horario = horario
        num_coinci = horarios_asignados[horario]

  #Una vez se tiene el mejor horario para el partido se asigna
  if num_coinci > 0: #Si ha habido una coincidencia se resta la audiencia correspondiente al partido previamente asignado
    repe = audiencias_finales[audiencias_finales["Horario"] == mejor_horario]
    audiencias_finales.loc[repe.index, 'Audiencia'] = audiencias_finales.loc[repe.index, 'Audiencia']*(1- reduccion_por_coincidencias2.iloc[num_coinci, 0])
    audiencia_tot =audiencia_tot - repe.loc[repe.index,'Audiencia'].iloc[0] + audiencias_finales.loc[repe.index, 'Audiencia'].iloc[0]
  audiencia_tot = audiencia_tot + mejor_audiencia
  horarios_asignados.loc[mejor_horario] = horarios_asignados.loc[mejor_horario] +1
  audiencias_finales.loc[partido] =[mejor_horario, mejor_audiencia]

print(audiencia_tot)
display(audiencias_finales)

```

```python
import random
import pandas as pd


# categorías de equipos y audiencias base
categorias = {'A': 3, 'B': 11, 'C': 6}
equipos_disponibles = ['A','A','A','B','B','B','B','B','B','B','B','B','B','B','C','C','C','C','C','C']
equipos_disponibles = random.sample(equipos_disponibles, len(equipos_disponibles))

audiencias_base = {('A', 'A'): 2, ('A', 'B'): 1.3, ('A', 'C'): 1, ('B', 'B'): 0.9, ('B', 'C'): 0.75, ('C', 'C'): 0.47}

#  horarios y sus ponderaciones
horarios_disponibles = ['V20', 'S12', 'S16', 'S18', 'S20', 'D12', 'D16', 'D18', 'D20', 'L20']
ponderaciones_horarios = {'V20': 0.4, 'S12': 0.55, 'S16': 0.7, 'S18': 0.8, 'S20': 1, 'D12': 0.45, 'D16': 0.75, 'D18': 0.85, 'D20': 1, 'L20': 0.4}

#audiencia
reduccion_por_coincidencias = {0: 0, 1: 0.25, 2: 0.45, 3: 0.6, 4: 0.7, 5: 0.75, 6: 0.78, 7: 0.8, 8: 0.8}

# Generar población inicialimport random

def generar_poblacion_inicial(tamano_poblacion, num_partidos, equipos_disponibles, horarios_disponibles):
    poblacion = []
    horarios = [horario for horario in horarios_disponibles if horario not in ['V20', 'L20']]
    for _ in range(tamano_poblacion):
        equipos = equipos_disponibles.copy()
        cromosoma = []
        for _ in range(num_partidos):
            equipo1 = random.choice(equipos)
            equipos.remove(equipo1)
            equipo2 = random.choice(equipos)
            equipos.remove(equipo2)
            # Asignar el horario obligatorio del Viernes o Lunes si aún no se ha asignado
            horario = 'V20' if 'V20' not in [partido[2] for partido in cromosoma] else 'L20'
            if horario not in [partido[2] for partido in cromosoma]:
                cromosoma.append((equipo1, equipo2, horario))
            else:
                cromosoma.append((equipo1, equipo2, random.choice(horarios)))
        poblacion.append(cromosoma)
    return poblacion


def calcular_aptitud(cromosoma, imprimir_resultado=False, iteracion=0):
    audiencia_total = 0
    horarios_ocupados = {}
    tabla = pd.DataFrame(cromosoma, columns=['Equipo1', 'Equipo2', 'Horario'])

    # Validar coincidencias de horario
    for partido in cromosoma:
        horario = partido[2]
        if horario in horarios_ocupados:
            horarios_ocupados[horario] += 1
        else:
            horarios_ocupados[horario] = 1

    i = 0
    # Calcular Audiencia
    for partido in cromosoma:
        equipo1, equipo2, horario = partido
        audiencia_base = audiencias_base.get((equipo1, equipo2), audiencias_base.get((equipo2, equipo1), 0))
        ponderacion_horario = ponderaciones_horarios[horario]
        base_ponderacion = audiencia_base * ponderacion_horario

        # Reducción de audiencia por coincidencias
        coincidencias = horarios_ocupados.get(horario, 0) - 1
        reduccion_porc = reduccion_por_coincidencias[coincidencias]
        reduccion_monto = base_ponderacion * reduccion_porc
        audiencia = base_ponderacion * (1 - reduccion_porc)
        audiencia_total += audiencia

        # Construir
        tabla.at[i,'Base Audiencia'] = round(audiencia_base,2)
        tabla.at[i,'Ponderacion Horario'] = round(ponderacion_horario,2)
        tabla.at[i,'Base * Ponderacion'] = round(base_ponderacion,2)
        tabla.at[i,'% Reduccion por Coincidencias'] = round(reduccion_porc,2)
        tabla.at[i,'Reduccion Audiencia'] = round(reduccion_monto,2)
        tabla.at[i,'Audiencia'] = round(audiencia,2)
        i += 1

    # Mostrar resultados finales
    if imprimir_resultado:
        tabla['Horario'] = pd.Categorical(tabla['Horario'], categories=horarios_disponibles, ordered=True)
        tabla = tabla.sort_values('Horario')
        tabla.insert(0, 'Partido', range(1, 1 + len(tabla)))
        tabla['Audiencia Final'] = tabla['Audiencia'].cumsum()
        print("Audiencia total:", audiencia_total)
        display(tabla)
    return audiencia_total


#  padres para el cruce
def seleccionar_padres(poblacion, num_padres):
    aptitudes = [calcular_aptitud(cromosoma) for cromosoma in poblacion]
    padres = random.choices(poblacion, weights=aptitudes, k=num_padres)
    return padres


# Cruzar
def contar_genes(cromosoma, gen):
    # Contar los genes en el cromosoma siempre y cuando estos no sean V20 o L20
    return sum(1 for alelo in cromosoma for g in alelo if g == gen)

def ajustar_cromosoma(cromosoma, conteos_deseados, genes_importantes):
    conteos_actuales = {gen: contar_genes(cromosoma, gen) for gen in conteos_deseados}
    diferencias = {gen: conteos_deseados[gen] - conteos_actuales[gen] for gen in conteos_deseados}

    if diferencias['A'] == 0 and diferencias['B'] == 0 and diferencias['C'] == 0:
        return cromosoma

    # Ajustar la cantidad de genes A, B y C
    for i, trio in enumerate(cromosoma):
        for j, gen in enumerate(trio[:2]):
            if diferencias[gen] < 0:
                # Buscar un gen para intercambiar
                for gen_reemplazo, diff in diferencias.items():
                    if diff > 0 and (j != 2 or trio[2] not in genes_importantes):
                        cromosoma[i] = trio[:j] + (gen_reemplazo,) + trio[j+1:]
                        diferencias[gen] += 1
                        diferencias[gen_reemplazo] -= 1
                        break


    return ajustar_cromosoma(cromosoma, conteos_deseados, genes_importantes)

def cruzar(cromosoma1, cromosoma2):
    punto_cruce = random.randint(1, len(cromosoma1) - 1)
    hijo1 = cromosoma1[:punto_cruce] + cromosoma2[punto_cruce:]
    hijo2 = cromosoma2[:punto_cruce] + cromosoma1[punto_cruce:]

    conteos_deseados = {'A': 3, 'B': 11, 'C': 6}
    genes_importantes = ['V20', 'L20']

    hijo1 = ajustar_cromosoma(hijo1, conteos_deseados, genes_importantes)
    hijo2 = ajustar_cromosoma(hijo2, conteos_deseados, genes_importantes)

    return hijo1, hijo2


# Mutar un cromosoma
def mutar(cromosoma, tasa_mutacion, equipos_disponibles, horarios_disponibles):
    equipos = equipos_disponibles.copy()
    # restringue horarios ya tomados de viernes y lunes = ['V20', 'L20']
    horarios = horarios_disponibles.copy()
    horarios = [horario for horario in horarios if horario not in ['V20', 'L20']]

    if random.random() < tasa_mutacion:
        for i in range(len(cromosoma)):
            equipo1 = random.choice(equipos)
            equipos.remove(equipo1)
            equipo2 = random.choice(equipos)
            equipos.remove(equipo2)

            # Asegurarse que L20 y V20 estén en el cromosoma
            if cromosoma[i][2] in ['V20']:
                horario = 'V20'
            elif cromosoma[i][2] in ['L20']:
                horario = 'L20'
            else:
                horario = random.choice(horarios)

            cromosoma[i] = (equipo1, equipo2, horario)

    return cromosoma




tamano_poblacion = [20]
num_generaciones = [40,50,60]
tasa_mutacion = [0.1,0.2,0.3]
num_partidos = 10

def algoritmo_genetico(tamano_poblacion, num_generaciones, tasa_mutacion, num_partidos, equipos_disponibles, horarios_disponibles):
    poblacion = generar_poblacion_inicial(tamano_poblacion, num_partidos, equipos_disponibles, horarios_disponibles)
    for _ in range(num_generaciones):
        nuevos_descendientes = []
        for _ in range(tamano_poblacion // 2):
            padres = seleccionar_padres(poblacion, 2)
            hijo1, hijo2 = cruzar(*padres)
            nuevos_descendientes.append(mutar(hijo1, tasa_mutacion, equipos_disponibles, horarios_disponibles))
            nuevos_descendientes.append(mutar(hijo2, tasa_mutacion, equipos_disponibles, horarios_disponibles))
        poblacion = nuevos_descendientes

    mejor_cromosoma = max(poblacion, key=calcular_aptitud)
    return mejor_cromosoma

def algoritmo_genetico_liga():
    iteracion = 0
    resultados = {}

    for i in tamano_poblacion:
        for j in num_generaciones:
            for k in tasa_mutacion:
                mejor_cromosoma = algoritmo_genetico(i, j, k, num_partidos, equipos_disponibles, horarios_disponibles)
                audiencia_total_mejor_cromosoma = calcular_aptitud(mejor_cromosoma, imprimir_resultado=True, iteracion=iteracion )

                print("Audiencia mejor:", audiencia_total_mejor_cromosoma)
                resultados[iteracion] = [mejor_cromosoma, audiencia_total_mejor_cromosoma, i, j, k, num_partidos]

                iteracion += 1


algoritmo_genetico_liga()
```
